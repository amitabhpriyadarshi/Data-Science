```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi,0.1)
print(x)
y=np.sin(x)
print(y)
plt.plot(x,y)
plt.show()
```

```python
#Create a 6x6 two-dimensional array, and let 1 and 0 be placed alternatively across the diagonals
x = np.zeros((6,6),dtype=int)
print("Array: x\n",x,"\n")

# y[:,:]   :,:  all rows ,all cols, start from 1st
y = np.zeros((6,6),dtype=int)
y[:,:]=1
print("Array: y\n",y,"\n")

# z[:2,:]   :2  first two rows start from 1st
z = np.zeros((6,6),dtype=int)
z[:2,:]=1
print("Array: z\n",z,"\n")

# a[::2,:]  : ::2 alternate rows start from 1st
a = np.zeros((6,6),dtype=int)
a[::2,:]=1
print("Array: a\n",a,"\n")

# b[::2,::2]  : [::2,::2] alternate rows and cols start from 1st
b = np.zeros((6,6),dtype=int)
b[::2,::2]=1
print("Array: b\n",b,"\n")

# c[1::2,:]  : 1::2 alternate rows start from 2nd skip row 1
c = np.zeros((6,6),dtype=int)
c[1::2,::2]=1
print("Array: c\n",c,"\n")

# alternate diagonal
d = np.zeros((6,6),dtype=int)
d[1::2,::2]=1
d[::2,1::2]=1
print("Array: d\n",d,"\n")

```

```python
# Find the total number and locations of missing values in the array
z = np.random.rand(4,4) # array of 10x10 with random number 0 to 1
z[np.random.randint(4,size=5),np.random.randint(4,size=5)] = np.nan # size =5 means five iterations

print(z)
print("\nTotal nulls in array:",np.isnan(z).sum())
print("\ncoordinates of nulls in array:\n",np.argwhere(np.isnan(z)))

indx = np.where(np.isnan(z))
print("\nindex of nulls in array:\n",indx)

z[indx] = 2
print("\n",z)
```
