#### Confusion Matrix
A confusion matrix is a table that is often used to describe the performance of a classification model (or "classifier") on a set of test data for which the true values are known.<br>
Confusion Matrix represents a tabular representation of Actual vs Predicted values.<br>You can calculate the accuracy of your model with:
<pre>
                  True Positives + True Negatives
___________________________________________________________________
True Positives + True Negatives + False Positives + False Negatives


    n:100   predicted YES | Predicted NO
________________________________________
Actual YES |    TP:49     |     FN:1
________________________________________
Actual NO  |    FP:5      |     TN:45

CF  = 49+45/(49+1+5+45)
    = 94/100
    = 0.94
    </pre>
<b>True Positives (TP):</b> These are cases in which we predicted yes (they have the disease), and they do have the disease.<br>
<b>True Negatives (TN):</b> We predicted no, and they don't have the disease.<br>
<b>False Positives (FP):</b> We predicted yes, but they don't actually have the disease. (Also known as a "Type I error.")<br>
<b>False Negatives (FN):</b> We predicted no, but they actually do have the disease.(Also known as a "Type II error.")<br>

#### Types of Probability

<b>1. Marginal Probability : </b>Marginal Probability is the probability of occurrence of a single event. Example: The probability that a card is of hearts</br>
<b>2. Joint Probability : </b>Joint Probability is a measure of two events happening at the same time. Example: The probability that a card is an Ace of hearts = P (Ace of hearts)</br>
<b>3. Conditional Probability : </b><br>
Probability of an event or outcome based on the occurrence of a previous event or outcome. <br>
Conditional Probability of an event B is the probability that the event will occur given that an event A has already occurred</br>
If A and B are dependent events then the expression for conditional probability is given by: <b><i>P(B|A) = P(A and B) / P(A)</i></b></br>
If A and B are independent events then the expression for conditional probability is given by <b><i>P(B|A) = P(B)</i></b></br>

Example

#### Hypothesis Testing
Statisticians use hypothesis testing to formally check whether the hypothesis is
accepted or rejected.<br>
<b>* Hypothesis testing is conducted in the following manner:<br>
State the Hypotheses - </b>This stage involves stating the null and alternative hypotheses.<br>
<b>* Formulate an Analysis Plan - </b>This stage involves the construction of an analysis plan. <br>
<b>* Analyse Sample Data - </b>This stage involves the calculation and interpretation of the test statistic as described in the analysis plan. <br>
<b>* Interpret Results - </b>This stage involves the application of the decision rule described in the analysis plan.
