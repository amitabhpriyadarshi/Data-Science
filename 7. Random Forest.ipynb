## Random Forest
<b>Classification</b> is the problem of identifying to which set of categories a new observation belongs.<br>
It is a <b>supervised learning model</b> as the classifier already has a set of classified examples and from these examples, the classifier learns to assign unseen new examples. Example: Assigning a given email into "spam" or "non-spam" category.<br>

<b><i>Types of Classifiers</i></b><br>
<b>Decision Tree</b>: Decision tree builds classification models in the form of a tree structure. It breaks down a dataset into smaller and smaller subsets.<br>
<b>Random Forest</b>: Random Forest is an ensemble classifier made using many decision tree models. Ensemble models combine the results from different models.<br>
<b>Naive Bayes</b>: It is a classification technique based on Bayes' Theorem with an assumption of independence among attributes.<br>

### Use Case - Credit Risk Detection

<b>Random Forest:</b> a versatile algorithm capable of performing both i) Regression ii) Classification. It is a type of ensemble learning method Commonly used predictive modelling and machine learning technique. it is an ensemble of decision trees.

### Use of Random Forest
Banking : Identification of loan risk applicants by their probability of defaulting payments.<br>
Medicine: Identification of at-risk patients and disease trends.<br>
Land Use: Identification of areas of similar land use.<br>
Marketing: Identifying customer churn.

### Random Forest Algorithm
<pre>
Step 1: Randomly select m features from T: where m«T 
Step 2: For node d, calculate the best split point among the m feature
Step 3: Split the node into two daughter nodes using the best split
Step 4: Repeat first three steps until n number of nodes has been reached 
Step 5: Build your forest by repeating steps 1-4 for D number of times
Step 6: Compile the results of all the D decision trees, and based on majority voting V we have our Random Forest.

Where: 
T is number of feature m, 
D is number of trees to be constructed, and 
V is Output of the class with the high vote.
</pre>
