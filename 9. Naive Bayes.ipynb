## Naive Bayes


### What is Naïve Bayes?
Naive Bayes is a simple but surprisingly powerful algorithm for predictive modeling.<br>


### Bayes Theorem and it's Use
Given a hypothesis H and evidence E , Bayes' theorem states that the relationship between the probability of the hypothesis before getting the evidence P(H) and the probability of the hypothesis after getting the evidence P(H|E) is
<pre>
P(H|E) = <u> P(E|H).P(H) </u>
              P(E)

Card Example: Probability of King in a deck of card

P(K) = 4/52 = 1/13

P(K|Face) = <u>P(Face|K)* P(K)</u>
                P(Face)
            
            P(Face|K) -> 1
            P(Face) -> 12/52 = 3/13
            
P(K|Face) = (1 * (1/13))/3/13
          = (1/13) / (3/13)
          = 1/3              
</pre>
### Mathematical working of Naïve Bayes
<pre>
P(A|B) = P(A∩B) /P(B)
P(B|A) = P(B∩A) /P(A)

P(A∩B) = P(A|B).P(B) = P(B|A).P(A)
       = P(A|B) = <u> P(B|A).P(A) </u>
                       P(B)
       
</pre>

### PIMA Diabetic Test

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

pima = pd.read_csv('./pima-indians-diabetes.csv')
pima.head()
```

```python
X= pima.iloc[:,:-1].values 
#[:,:-1] = first : takes every row. :-1 select every column expect last one
y= pima.iloc[:,8].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=0)
print(len(X_train))
print(len(X_test))
```

```python
#fitting multiple linear rigression to the training set
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cf_matrix = confusion_matrix(y_test, y_pred)
print(cf_matrix)

score = accuracy_score(y_test, y_pred)
print(score)
```
