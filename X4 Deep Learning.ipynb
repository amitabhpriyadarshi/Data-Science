# 14. Deep Learning
### What Is Deep Learning?
Humans have seen so many dogs and cats, they can identify the difference between the two.<br> machine learning algorithms too needed our help to point out the features for the model to learn and predict.<br>
But its not possible for manually extraction of features. so, the Deep learning is born.<br>

<b>Deep Learning skips the manual steps of extracting features, you can directly feed images to the deep learning algorithm, which then predicts the object.</b>


### How Deep Learning Works?
Deep learning is a form of machine learning that uses a model of computing that's very much inspired by the structure of the brain. 
A type of artificial neuron is called a perceptron.


Step 1: Initialialize the weights and threshold<br>
Step 2: Provide the input and calculate the output<br>
Step 3: Update the weights. new weight = old weight + learning rate<br>
Step 4: repeat step 2 and 3
